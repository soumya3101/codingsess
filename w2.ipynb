Now lets move on to **For loops!**

A for loop is simply used for iterating over a sequence. With the for loop we can execute a set of statements, once for each item in a list, a set etc. Now lets go through a simple for loop. 

Here I am going to print my favorite fruits in order. 

```python
fruits = ['apple', 'orange', 'pineapple', 'watermelon', 'avacado'] #recall lists from the previous week

for x in fruits:
    print(x)
#this loop is simply taking each element in the list 'fruits' and printing them in order. In the first iteration x is going 
#to be equal to apple since its the first in the list of 'fruits'. Then the loop prints the x and moves on to the second
#iteration. THe second iteration selects orange and print it and moves on to the third. The for loop runs until x is equal to 
#the last element in the list 'fruits'
```

```python
#now lets try a bit more complicated for loop
x = np.zeros([5]) # i am creating a numpy array with 5 zeros. The goal of the loop is to fill this array x with ones using 
#a for loop. Now we want the loop to iterate 5 times. So we can use the range(0,5). This will use a range of numbers from 0 to 4
#inclusive

for i in range(0,5):
    x[i] = 1
print(x)
```

As expected the matrix x was filled with ones.
